```python
from transformers import AutoTokenizer

MODEL_ID = "google/gemma-3-1b-pt"
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)

# Get the newline token ID
newline_id = tokenizer.encode("\n", add_special_tokens=False)[0]
print(f"Newline token ID: {newline_id}")
print(f"Newline token decoded: {repr(tokenizer.decode([newline_id]))}")

# Also show what the tokenizer thinks about newline
print(f"\nTokenizer vocab lookup for newline: {tokenizer.convert_tokens_to_ids(['\\n'])}")

# Show the actual token string representation
print(f"Token at ID {newline_id}: {repr(tokenizer.convert_ids_to_tokens([newline_id]))}")
```
